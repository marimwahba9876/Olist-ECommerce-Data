```python
def check_null_products_in_orders(
        products,order_items,check_columns,id_columns = "product_id"):
         null_products = products[products[check_columns].isnull()]
         null_ids = null_products[id_columns]
         matching_orders = order_items[order_items[id_columns].isin(null_ids)]
         print(f"===column report: {check_columns}===")
         print(f"number of products that contain null in products table: {len(null_products)}")
         print(f"number of times these products appeared in orders tables: {len(matching_orders)}")
         if len(matching_orders)> 0:
                 print("these products are sold do not drop them")
         else:
                 print("these products are not sold you can drop them")
         return matching_orders        
    
```

```python
columns_to_check = ["product_weight_g","product_length_cm","product_height_cm","product_width_cm","product_category_name","product_name_lenght","product_description_lenght","product_photos_qty"]
for col in columns_to_check:
    check_null_products_in_orders(products, order_items,check_columns=col)
```

# Section 2: Data Cleaning

## Order table

```python
## data type conversion
datetime_col = ["order_purchase_timestamp",
                "order_approved_at",	
                "order_delivered_carrier_date",
                "order_delivered_customer_date",	
                "order_estimated_delivery_date"]
for col in datetime_col:
    orders[col] = pd.to_datetime(orders[col])
```

```python
orders.dtypes
```

## order_items (FACT TABLE)

```python
order_items["shipping_limit_date"] = pd.to_datetime(order_items["shipping_limit_date"])
```

```python
order_items.dtypes
```

## order_reviews (FACT TABLE)

```python
reviewdate_col = ["review_creation_date",
                  "review_answer_timestamp"]
for cols in reviewdate_col:
    order_reviews[cols] = pd.to_datetime(order_reviews[cols])
```

```python
order_reviews.dtypes
```

```python
order_reviews["review_comment_message"] = order_reviews["review_comment_message"].fillna("customer did not leave a comment")
order_reviews["review_comment_title"] = order_reviews["review_comment_title"].fillna("review not provided")
```

## Customers

```python
## convert dtype of customer_zip_code_prefix fro int64 to string
customers["customer_zip_code_prefix"] = customers["customer_zip_code_prefix"].astype(str)
```

```python
customers.dtypes
```

```python
## Convert all letters to lowercase
customers["customer_city"] = customers["customer_city"].str.lower().str.strip()
```

## Products

```python
## replace null values in category name with "unknown"
```

```python
products["product_category_name"] = products["product_category_name"].fillna("unkown")
```

```python
products.info()
```

```python
# fill null values in "product_name_lenght","product_description_lenght","product_photos_qty" with 0
products_t1_nullvalues = ["product_name_lenght","product_description_lenght","product_photos_qty"]

for p in products_t1_nullvalues:
    products[p] = products[p].fillna(0)
```

```python
products.describe()
```

```python
# fill null values in product_weight_g	product_length_cm	product_height_cm	product_width_cm with the median of each column
products["product_weight_g"] = products["product_weight_g"].fillna(products["product_weight_g"].median())
products["product_length_cm"] = products["product_length_cm"].fillna(products["product_length_cm"].median())
products["product_height_cm"] = products["product_height_cm"].fillna(products["product_height_cm"].median())
products["product_width_cm"] = products["product_width_cm"].fillna(products["product_width_cm"].median())
```

```python
products.info()
```
